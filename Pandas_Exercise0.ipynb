Clean Employee Data:
create employee dataset with empid,name,salary,years of experience,department with duplicate values,null values.


```python
import numpy as np
import pandas as pd
from google.colab import files

uploaded=files.upload()
```

```python
df=pd.read_csv("employee_dataset.csv")
print(df)
```

```python
#Find the missing values in each column
df.isnull().sum()
```

```python
# Replace missing names with unknown.
df["Name"].isnull().sum()
```

```python
# Fill Missing Departments with the most frequent department.
df["Department"].fillna(df["Department"].mode()[0], inplace=True)
print(df)
```

```python
#replace missing salaries with the median salary
df['Salary'].fillna(df["Salary"].median(),inplace=True)
print(df)
```

```python
#Replace misssing experience with the average experience
df.fillna(df['Years_of_Experience'].mean(),inplace=True)
print(df)
```

```python
#Rename column to lowercase
df.columns=df.columns.str.lower()
print(df)
```

```python
#Remove duplicates
dup=df.drop_duplicates(subset="name")
print(dup)
```

```python
#Display all exployees from the it department
df[df["department"]=="IT"]
```

```python
#Find employees whose salary is greater than 70000
df[df["salary"]>70000]
```

```python
#Show employees working in hyderabad
df[df['city']=='Hyderabad']
```

```python
#Find experience with more than 5yers of experience
df[df['years_of_experience']>5.0]
```

```python
#Show employees who belong to either IT or Finance department
df[(df['department'].isin(['IT','Finance']))]
```

```python
#Display employees whose salary lies between 60000 and 80000
df[df["salary"].between(60000,80000)]
```

```python
#Find the number of employees in each department
df.groupby("department")["name"].count()
```

```python
#Calculate total salary paid by each department
df.groupby("department")["salary"].sum()
```

```python
#Find the average salary of employees in each city
df.groupby("city")["salary"].mean()
```

```python
#Using a single aggregation,calculate for each department:
#avg,max,min,total salary
df.groupby("department")["salary"].agg(["mean","max","min","sum"])
```

```python
# Create a pivot table showing average salary by department
pd.pivot_table(df,values="salary",index="department",aggfunc="mean")
```
