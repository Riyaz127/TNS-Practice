Check whether a Number is Armstrong Number?

```python
num=int(input("Enter a Number:"))
temp=num
order=len(str(num))
sum_value=0
while temp>0:
      digit=temp%10
      sum_value +=digit**order
      temp //=10
if sum_value==num:
  print("It is an Armstrong Number.")
else :
  print("It is not a Armstrong Number.")
```

Find LCM of two Numbers

```python
a=int(input("Enter a Number:"))
b=int(input("Enter a Number:"))
greater=max(a,b)
while True:
  if greater%a==0 and greater % b==0:
    lcm=greater
    break
  greater += 1
print("lcm: ",lcm)
```

Prime Numbers Between 1 & 100

```python
for num in range(1,101):
  if num >1:
    for i in range(2,int(num**0.5)+1):
      if (num % i)==0:
        break;
    else:
      print(num,"")
```

Count Vowels and Constants in a String?

```python
word=input("Enter a String.")
vowels="aeiouAEIOU"
vowel_count=0
consonent_count=0
for i in word:
  if i in vowels:
    vowel_count +=1
  else:
    consonent_count+=1
print("vowels",vowel_count)
print("consonent",consonent_count)
```

Largest Element in a list without using Max()?

```python
a=[23,6,78,6,12,19,37,40]
largest=a[0]
for i in a:
  if i >largest:
    largest=i
print("largest Number in the list: ",largest)
```

Remove Duplicate elements from a list?

```python
a=[23,78,45,23,56,89,56,74,45]
unique_list=[]
for i in a:
  if i not in unique_list:
    unique_list.append(i)
print(unique_list)
```

Generate Patterns using nested loops?

```python
for i in range (0,6):
  for j in range(1,i+1):
    print("*",end='')
  print()
```

Print Pascal's Triangle ?

```python
for i in range (0,6):
  for j in range(6,i+1):
    print("*",end='')
  print()

  for k in range(i-1,6):
    print("*",end='')
  print()
```
