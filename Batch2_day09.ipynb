```python
#which city has the highest average marks
city_avg=df.groupby("City")["Marks"].mean()
print("Average Marks by city:",city_avg)
highest_city=city_avg.idxmax()
print("\n City with the highest average Marks:",highest_city)
```

```python
#who is the topper in each class
topper=df.loc[df.groupby("Class")["Marks"].idxmax()]
print(topper[["Class","Name","Marks"]])
```

```python
#how many students failed?(pass marks>40)
failed=df[df['Marks']<40]
print(failed.value_counts())
```

```python
# overall pass percentage
total=len(df)
pass_percent=len(df[df["Marks"]>40])
percent=(pass_percent/total)*100
print(round(percent,2))
```
