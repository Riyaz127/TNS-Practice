```python
import numpy as np
import pandas as pd
from google.colab import files

uploaded=files.upload()
```

```python
df=pd.read_csv("student.csv")
print(df)
```

```python
#Filtering students who scored marks > 80
print(df[df['Marks']>80])
```

```python
# Sorting sort the df based on marks in descending
df.sort_values(by="Marks",ascending=False)
```

```python
#Adding Columns
def asign_grade(score):
  if score >=90:
    return "A"
  elif score>=80 and score<90:
    return "B"
  elif score>=70 and score<80:
    return "C"
  else:
    return "D"
df["Grade"]=df["Marks"].apply(asign_grade)
print(df)
```

```python
#modifying column
# convert all the names of the students into upper case letters
upper_case=df["Name"].str.upper()
print(upper_case)
```

```python
#Removing columns
#remove department column from dataframe
df.drop(columns='Department')
```

```python
#Dropping rows
#remove the student records index 3
df.drop(index=3)
```

```python
#Groupby
#find the average marks in each department
df.groupby("Department")["Marks"].mean().round(2)
```

```python
#Aggregate Function
#calculate the mean,max,min,sum,median,std.(statistics)
statistic=df["Marks"].agg(["mean","max","min","sum","median","std"])
print(statistic)
# stat=df.groupby("Department")["Marks"].agg(["mean","max","min","sum","median","std"])
# print(stat)
```

```python
#Combining DataFrame
#create an another dataframe and add both the dataframe
new={
    "Name":["Teja","Divya"],
    "Age":[20,21],
    "Marks":[87,91],
    "Department":["CSE","ECE"],
    "Exam_date":['2026-06-20','2026-06-21']
}
df1=pd.DataFrame(new)
df1["Grade"]=df1["Marks"].apply(asign_grade)
# print(df1)
#Contatination
data=pd.concat([df,df1])
print(data)
```

```python
from datetime import date
#Working with data
# convert exam_date format and extract the day,month,year
date_time=pd.to_datetime(df["Exam_date"])
yr=pd.to_datetime(df["Exam_date"]).dt.year
mn=pd.to_datetime(df["Exam_date"]).dt.month
dy=pd.to_datetime(df["Exam_date"]).dt.day
print(date_time)
print("year\n",yr)
print("Month\n",mn)
print("Day\n",dy)
```

```python
# Indexing
# set the name column as the index and display the details of the student "Rahul"
print(df[df['Name']=="Rahul"])
```

```python
#Pivot Table
#Create a pivot table showing the average marks for each department
pd.pivot_table(
    data=df,
    index="Department",
    values="Marks",
    aggfunc=("mean","sum")
)
```
