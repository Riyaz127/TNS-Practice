Explonatory Data Analysis(EDA)

```python
import pandas as pd
data={
    "Exam_Score":[55,65,72,60,95,80,85,250]
}
df=pd.DataFrame(data)
Q1=df['Exam_Score'].quantile(0.25)
Q3=df['Exam_Score'].quantile(0.75)
IQR=Q3-Q1
Lower=Q1-1.5*IQR
Upper=Q3+1.5*IQR
# print(lower)
# print(upper)
outliers = df[
    (df["Exam_Score"] < Lower) | (df["Exam_Score"] > Upper)
    ]
print(outliers)

```

Find the Outliers for below data using IQR python code:


```python
# 1. age=[22,23,24,25,26,27,28,29,120]
Age={
"age":[22,23,24,25,26,27,28,29,120]
      }
da=pd.DataFrame(Age)
q1=da['age'].quantile(0.25)
q3=da['age'].quantile(0.75)
iqr=q3-q1
lower=q1 - 1.5 * iqr
upper=q3 + 1.5 * iqr
Outlier=da[
    (da['age']<lower)|(da['age']>upper)
]
print(Outlier)
```

```python
#2. data=[5,6,7,8,9,10,11,12,100]
Age={
"Data":[5,6,7,8,9,10,11,12,100]
      }
da=pd.DataFrame(Age)
q1=da['Data'].quantile(0.25)
q3=da['Data'].quantile(0.75)
iqr=q3-q1
lower=q1 - 1.5 * iqr
upper=q3 + 1.5 * iqr
Outlier=da[
    (da['Data']<lower)|(da['Data']>upper)
]
print(Outlier)
```
