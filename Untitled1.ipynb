```python
# Stack implementation
class Stack:
    def __init__(self):
        # Create an empty list to store stack elements
        self.items = []

    # Add an element to the top of the stack
    def push(self, item):
        self.items.append(item)

    # Remove and return top element
    def pop(self):
        if self.is_empty():
            print("Stack is empty")
            return None
        return self.items.pop()

    # Return the top element
    def peek(self):
        if self.is_empty():
            print("Stack is empty")
            return None
        return self.items[-1]

    # Check if stack is empty
    def is_empty(self):
        return len(self.items) == 0

    # Check stack size
    def size(self):
        return len(self.items)

    # Display stack elements
    def display(self):
        print(self.items)

# Create stack object
stack = Stack()

# Push elements onto the stack
stack.push(30)
stack.push(67)
stack.push(44)
stack.push(98)

# Display stack
stack.display()

# View top element
print("Top:", stack.peek())
```
