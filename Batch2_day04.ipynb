```python
# Write a function to reverse a string?
def reverse(str):
  str1=""
  for i in str:
    str1=i+str1
  return str1
str= input("Enter a Word.")
print("Reversed String: ",reverse(str))
```

```python
#write a function to know a number is a Armstrong of not?
Num = int(input("Enter a Number: "))

def armstrong(Num):
    temp = Num
    order = len(str(Num))
    total = 0

    while temp > 0:
        digit = temp % 10
        total += digit ** order
        temp //= 10

    if total == Num:
        print("It is an Armstrong Number.")
    else:
        print("It is not an Armstrong Number.")

armstrong(Num)


```

```python
# Write a function to generate fabinocci series
def fabinocci(n):
  a,b=0,1
  for i in range(n):
    print(a,end=" ")
    a,b=b,a+b

num=int(input("Enter Length of the series."))
fabinocci(num)
```

```python
# write a function to find the sum of digits of a number.
num=int(input("Enter a Number:"))
def sum_of_digits(n):
  temp=n
  sum=0
  while temp>0:
    digit= temp%10
    sum += digit
    temp //  10
  print(sum)
sum_of_digits(num)
```
