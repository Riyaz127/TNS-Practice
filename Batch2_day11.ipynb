```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("supermarket_dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df["City"].fillna(df["City"].mode()[0],inplace=True)
df["Price"].fillna(df["Price"].mean(),inplace=True)
df['Quantity'].fillna(df['Quantity'].mean(),inplace=True)
df["Customer_Name"].fillna(df["Customer_Name"].mode()[0],inplace=True )
```

```python
df.isnull().sum()
```

```python
print(df)
```

```python
# Line Plot
plt.plot(df.groupby('Payment_Method')['Price'].sum())
plt.xlabel("Payment Mode")
plt.ylabel("Price")
plt.title("Payment Mode Vs Price")
plt.show()
```

1. Why did you choose this graph?                                       
Ans: which payment mode has beeing generating more revenue.
2. What business question does it answer?                            
Ans:Customers are interested in which payment mode.
3. What insight did you obtain?                                      
Ans: Most of the customers are willing to pay through UPI so Try to upgrade UPI methods and Eliminate card Methods.


```python
#Bar plot
plt.figure(figsize=(10,5))
plt.bar(df['Product'],df['Quantity'])
plt.xlabel("Products")
plt.ylabel("Quantity")
plt.title("Orders by Category")
plt.show()
```

1. Why did you choose this graph?                                       
Ans: To compare the number of orders by different category.
2. What business question does it answer?                            
Ans:Which product has more demand in the market.and less demand.
3. What insight did you obtain?                                      
Ans: Maintainance of the demanded product.


```python
#Scatter Plot
plt.scatter(df['Quantity'],df['Price'])
plt.xlabel("Quantity")
plt.ylabel("Price")
plt.title("Price Vs Quantity")
plt.grid()
plt.show()
```

1. Why did you choose this graph?                                       
Ans: To Know the relation between two numerical variables.
2. What business question does it answer?                            
Ans:does the price affect the quantity order by customer.
3. What insight did you obtain?                                      
Ans: Quantity and price are inversely proportional
.
