```python
# Decision Tree (CLASSIFICATION AND REGRESSION)
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

df=pd.DataFrame({
    "Income":[25,35,45,60,70,80],
    "Loan":[0,0,0,1,1,1]
})
X=df[["Income"]]
y=df["Loan"]

model=DecisionTreeClassifier()
model.fit(X,y)

prediction=model.predict([[65]])
print("Loan Approved: ",prediction[0])
```

```python
# Logistic Regression (CLASSIFICATION)
import pandas as pd
from sklearn.linear_model import LogisticRegression

data=pd.DataFrame({
    "Hours":[1,2,3,4,5,6],
    "Pass":[0,0,0,1,1,1]
})
X=data[["Hours"]]
y=data["Pass"]

model=LogisticRegression()

model.fit(X,y)

predicted=model.predict([[6.5]])
print(predicted[0])
if predicted[0]==1:
  print("Passed")
else:
  print("Failed")
```

```python
#Simple Linear Regression (Regression)
import pandas as pd
from sklearn.linear_model import LinearRegression

df=pd.DataFrame({
    "Expenses":[1,2,3,4,5],
    "Sales":[3,4,5,6,7]
})
x=df[["Expenses"]]
y=df["Sales"]

model=LinearRegression()
model.fit(x,y)
prediction=model.predict([[6]])
print("Sales Predicted:",prediction[0])
print("Slope",model.coef_[0])
print("Intercept",model.intercept_)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df=pd.DataFrame({
    "Area":[1000,1200,1500,1800,2000],
    "Bedrooms":[2,2,3,3,4],
    "Age":[10,8,5,2,1],
    "Price":[40,45,60,75,90]
})
X=df.drop('Price',axis=1)
Y=df["Price"]

model=LinearRegression()
model.fit(X,Y)

Prediction=model.predict([[1700,4,15]])

print("Price: ",Prediction[0])
```
